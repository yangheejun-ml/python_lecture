# 이론 활용 및 실전 프로그래밍 안내

## 오늘 시작한 사람은 어제 시작한 사람보다 늦습니다.

우리는 불과 10초전에 미국에 무슨일이 일어났었는지를 알 수 있는 정보의 시대에 살고 있습니다. 검색만 해보면 그리고 내가 그걸 검색할 능력만 있으면 내가 원하는 대부분의 자료는 이미 인터넷에 있습니다. 혹자는 그래서 우리에게 그만큼의 기회가 없는 세상을 살고 있다고 합니다. 이미 없는게 없을만큼 다 나와있기 때문에 새로운게 없다는 의미입니다.


오늘 시작한 우리는 어제 시작한 그들보다 늦을 수 밖에 없습니다. 그래서 어제 그들이 만들어놓은 수많은 프로그램들을 사용하면서 오늘을 보내고 있게 됩니다. 이런 상황에서 우리는 항상 새로운 무언가를 찾는다는게 어제 그들보다 어렵다고 생각하게 됩니다. 하지만 이게 사실 맹점인게 어제 그들의 노력으로 인해 그들이 제공하는 많은 자료를 활용 할 수 있는 강점이 생기게 됩니다. 


제 개인적인 생각의 4차 산업혁명의 시대는 사실 기술의 시대가 지나고 이제 아이디어의 시대라고 보여집니다. 우리가 현재 알고 있는것보다 인공지능은 생각보다 많은 발전을 이루고 있습니다. 인간을 대처할 수 있는 수많은 분야에서 이미 인공지능은 자리를 잡아가고 있는 실정입니다. 그러면 이런 인공지능의 시대에 인간의 포지션은 이런 인공지능이 하지 못하는 아이디어의 영역이어야 한다고 생각합니다.

물론 인공지능이 작곡을 하고 영화를 만들고 소설을 쓸 수 있는 시대이지만 그런 인공지능을 만들어낸 아이디어는 인간에게 나옵니다. 

## 아이디어

창의력은 수많은 경험과 정보를 기반으로 그것이 유기적인 융화과정을 거쳤을때 새로운 아이디어가 탄생한다고 봅니다. 말이 어려운데 쉽게 경험과 지식이 많은 사람에게 더 좋은 아이디어가 나온다 뭐 그런 단순한 얘기입니다. 우리가 주변에 사용하고 있는 혁신적인 모든것들은 그것을 만든 사람들이 처음부터 창조한건 없습니다. 다 그 이전 세대의 정보를 활용하고 가공하여 조금 더 편리하게 진화시켰을 뿐입니다. 어쨌든 그런 무언가는 최초에 시작이라는 시작점이 있어야 하는데 그게 사실 참 어려울때가 많습니다.

뭔가를 해보고 싶은데 뭘 어디서부터 어떻게 해야할지를 찾는다는게 참 어렵습니다. 그래서 우리는 어제 시작한 그들의 지식을 최대한 활용해서 그걸 보다 편리하게 진화시킬 수 있습니다.

### 따라 해보는 실습 위주의 프로그래밍 연습

물론 파이썬의 전부를 모두 다 다룬것은 아니지만 기본적으로 필요한 내용은 거의 다 공부했으니 이제 그걸 써먹어야 할 차례 입니다. 이번 강의부터는 실제 하나의 목적이 뚜렷한 주제를 정하고 그걸 프로그래밍 하는 연습을 합니다. 프로그래밍의 목적은 문법이 어떻고 저떻고 함수가 어떻고 저떻고 하는게 아니라 프로그램을 만들어내는게 목적입니다. 

### 완성된 소스코드를 보는것과 그 소스코드를 만들어 가는 과정을 보는건 전혀 다른 이야기 입니다.

타인이 만든 프로그램의 완성된 코드를 보는것과 그걸 만들어 가는 과정을 보는건 전혀 다른 이야기 입니다. 

*어떤 간단한 프로그램의 전체 소스코드를 보고 100% 이해했다고 하더라도 막상 그 프로그램을 만들어보라고 하면 대부분 못 만듭니다.*

어떤 순서로 어떻게 프로그램이 만들어지는지를 볼 수 있다는건 하나의 커다란 기회 입니다. 그 사람이 A 함수를 먼저 만들었는지 B 함수를 먼저 만들었는지 A함수를 먼저 만들었다면 왜 그걸 먼저 만들게 되었는지, 이게 결국 로직의 흐름이 되고 로직을 짜보는 경험이 많을 수록 점점 좋은 결과물이 나올겁니다. 물론 정해진 정답은 없지만 그렇기 때문에 이런 실습 위주의 프로젝트는 정말 중요한 기회 입니다.

* 요리를 글로 배운사람과 실습을 해본 사람의 차이

### 이론만으로는 절대 프로그래밍은 늘지 않습니다. 

위에서 얘기한대로 타인이 만들어놓은 소스코드를 100% 이해했다고 해도 직접 만들어보라고 하면 못 만드는 이유와 같습니다. 이해하는것과 활용해서 만드는건 다른 이야기 입니다. 눈으로 보고 이해했다고 해도 한번 직접 코딩을 해보는것과는 천지차이 입니다.

물론 늘 시간에 쫓기는 실무에선 코드를 복사하고 붙여넣기 하는 경우가 많습니다. 하지만 본인이 초보 개발자라고 생각한다면 아무리 간단한 로직의 코드라도 직접 코딩하세요. 대충 눈으로 보고 "아~ 이런 코드구나 쉽네~" 라고 생각만 하는 사람과 그 쉬운걸 한번이라도 코딩해보는 사람이 있다면 장담하건데 후자의 경우가 결국엔 좋은 개발자가 될 겁니다. 

프로그램을 공부할때는 이론을 공부하지만 실제 프로그래밍을 하는건 그 이론을 활용해서 로직을 만들어내는 것 입니다. 아무리 이해하고 있다고 생각해도 활용해서 로직을 만들어내기까지는 그 환경에 익숙해지는 시간이 필요합니다. 무언가에 익숙해지는건 많이 해볼수록 빠릅니다. 익숙해지지 않은걸 잘 할 수 있는 방법은 없습니다. 간단한 프로그램이라도 내가 직접 코딩해보고 내가 직접 결과를 보는게 중요합니다. 무조건 코딩해보세요 그게 정답입니다.

### 많은 분들이 쉽게 포기합니다.

100명이 프로그래밍 공부를 시작하면 결국엔 진짜 개발자는 몇명이나 될 수 있을까요? 물론 제가 통계를 내본건 아니라서 정확하게 얘기할 수 없지만 아마도 몇 명 안되는 극소수만 남을겁니다. 나머지는 모두 포기하게 됩니다. 물론 프로그래밍이라는게 어느정도 적성에 맞지 않으면 쉽지 않은 일인건 사실입니다. 우리 모두가 운동선수가 될 수 없는 이유와 같은 이유겠죠. 어쨌든 어느정도 적성에 맞지 않으면 하기 힘든 일이 있습니다. 그렇긴하지만 막상 실제 포기하는 경우는 그런 적성의 문제보다 계획의 문제인 경우가 더 많습니다.

보통 우리는 인터넷 같은 정보속에서 최상의 결과만을 보고 살아갑니다. 그리고 거기에 맞게 눈이 높아져 있습니다. 그러나 실상 내 자신의 실력은 상대적으로 뒤쳐집니다. 그게 사실 당연한건데 그 당연한걸 못 받아들일 때가 많습니다.

어느날 인터넷을 보다가 레고로 만든 멋진 작품을 보았습니다. 대충 검색해보고 알아보니 만들어놓은걸 보니 할만 해 보입니다.
레고 조립을 한번도 안해본 사람이 3천피스가 넘는 레고조립을 시도합니다. 그리고 어렵다고 합니다. 그러다 나 스스로를 의심하거나 여러가지 합리화를 통해 포기합니다. 당연한 얘기가 아닐까요? 아주 작은 집부터 만들어보는걸 추천합니다. 그렇게 만들다보면 어느날엔가는 성을 만들어보기도 하고 새로운 아이디어가 생기는 날이 올 거라 생각합니다.

공부만 열심히 해서 이론만 아무리 빠삭하게 알고 있는 사람과 매일 뭘 만들어보고 똑같은걸 또 다르게 만들어보고 하는 사람은 시간이 지날수록 격차가 벌어집니다. 결국엔 많이 만들어본 사람이 실무에서 개발일을 하는 경우가 많습니다. 개인적인 생각이지만 실무 경험에서 하는 이야기 입니다.

### 오류랑 친해지기

프로그래밍을 하다보면 수많은 오류와 난관에 부딪히는 경우가 많고 그게 정상입니다. 개발은 오류와 예외처리와의 싸움입니다. 컴퓨터의 환경은 저마다 다릅니다. 그래서 여기서는 잘 되는데 다른데선 안되는 경우도 많습니다. 본인 스스로가 해결하고자 하는 의지가 없다면 절대 발전 할 수 없는 영역입니다.

그리고 오류와 빨리 친해질수록 발전하는 속도도 빠릅니다. 오류를 걱정하거나 두려워 하지 마시길 바랍니다. 오래 걸리는건 상관없습니다. 내가 발전할 수 있는 기회이고 반드시 해결해야하는 문제일 뿐입니다.

그런 오류를 만났을때 구글을 활용하여 오류를 해결하는 습관을 들이는게 가장 중요합니다. 구글이 있는 방대한 다양한 국적의 사람들이 이미 그걸 고민했었고 해결한 사람도 있고 힌트를 얻을 수 있습니다.
