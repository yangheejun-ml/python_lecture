## 최종 코드 완성

```python
# tkinter 라이브러리
import tkinter
import socket
from threading import Thread

# 사용자가 접속하기를 누르지 않고 그냥 접속입력창을 닫았을시 프로그램을 종료하기 위해 사용
import sys

# 접속되 서버의 IP 와 포트 번호 변수
IP = ""
PORT = 0

def connect(event=None):
    '''접속하기 버턴을 클릭하면 호출 되는 함수
    tkinter.Button 생성시 command= 인자에 함수명을 기재
    '''
    # 이 함수에서 사용하는 변수는 외부에 선언된 변수를 사용함을 명시적으로 선언함
    global IP, PORT
    
    # StringVar 객체에서 str() 문자열 값을 얻어냅니다.
    input_string = input_addr_string.get()

    # 127.0.0.1:13000 값일때 : 를 기준으로 나눕니다.
    addr = input_string.split(":")
    # 0번째는 아이피 1번째는 포트번호 입니다.

    IP = addr[0]
    PORT = int(addr[1])

    print("서버 접속 [{}:{}]".format(IP, PORT))

    # 서버 정보 입력 윈도우는 파기
    win_input.destroy()

    
def recv_message():
    '''서버에서 채팅 데이터를 수신하는 함수'''
    
    while True:
        # 서버로부터 1024 만큼 데이터 수신
        msg = sock.recv(1024)

        # 채팅 목록 리스트 위젯 마지막줄에 수신된 데이터를 utf-8로 디코딩 후 삽입
        chat_list.insert(tkinter.END, msg.decode("utf-8"))

        # 채팅 목록 리스트를 맨 마지막으로 스크롤
        chat_list.see(tkinter.END)
        
    
def send_message(event=None):
    '''채팅 내용 입력 후 엔터버튼 혹은 전송하기 클릭시 서버로 채팅 데이터를 전송할 함수'''
    
    # input_msg 는 채팅입력 위젯과 변수를 연결하는 객체 입니다.
    # str 데이터를 받기 위해선 get() 함수를 사용합니다.
    message = input_msg.get()
    
    print("[전송]: {}".format(message))
    # 채팅 내용을 utf-8 bytes 형태로 형변환 해서 서버로 전송
    sock.send(bytes(message, "utf-8"))
    
    # 기존의 채팅내용은 초기화
    input_msg.set("")
    
    # 종료 명령 확인
    if message == "/bye":
        sock.close()
        window.destroy()
        sys.exit(1)

        
def window_input_close(event=None):
    '''사용자가 서버 접속정보 입력창에서 X 버튼을 클릭해 창을 닫은경우 호출되는 함수'''
    print("윈도우 종료")
    # 윈도우를 제거 합니다.
    win_input.destroy()
    # 프로그램을 종료 합니다.
    sys.exit(1)

    
def window_chat_close(event=None):
    '''사용자가 채팅창에서 X 버튼을 클릭해 창을 닫은경우 호출되는 함수
    서버에 종료 사실을 알려야 하기 때문에 그냥 윈도우를 종료하지 않고
    서버에 /bye 명령을 날리고 전송해야 합니다.
    '''
    input_msg.set("/bye")
    send_message()
    
# 입력창을 만들 tkinter 객체를 생성 합니다.
win_input = tkinter.Tk()


# 사용자가 접속정보 입력창 윈도우에서 X 버튼을 클릭했을때
# 접속 의사가 없다고 판단하고 프로그램을 종료 하기 위해
# X 버튼 클릭시 발생하는 WM_DELETE_WINDOW 메세지를 window_input_close() 함수와 연결합니다.
win_input.protocol("WM_DELETE_WINDOW", window_input_close)

# 윈도우 타이틀이름을 설정합니다.
win_input.title("접속대상")

# 맨 왼쪽 접속대상 라벨 박스를 만듭니다.
# 한 줄에 라벨, 입력박스, 버튼 이렇게 3개의 위젯을 배치하기 위해서
# tkinter 의 grid 함수를 사용해 위치를 배정해 줍니다.
tkinter.Label(win_input, text="접속대상").grid(row=0, column=0)

# 입력박스(Entry) 로 입력받은 값을 변수로 사용하기 위해
# tkinter 의 StringVar 로 객체를 생성했고 기본값으로 "127.0.0.1:13000" 값을 주었습니다.
input_addr_string = tkinter.StringVar(value="127.0.0.1:13000")

# 입력박스의 컨트롤 객체를 생성합니다.
# 인자값으로 부모윈도우 변수, StringVar 변수명, width 값을 설정했습니다.
input_addr = tkinter.Entry(win_input, textvariable=input_addr_string, width=20)

# grid() 함수를 사용해서 위젯의 위치를 배정하고 상하좌우 여백을 5 로 설정했습니다.
input_addr.grid(row=0, column=1, padx=5, pady=5)

# 버튼 객체를 생성하고 클릭했을시 connect 라는 함수와 연결합니다.
connect_button = tkinter.Button(win_input, text="접속하기", command=connect)

# grid() 함수를 사용해서 위젯의 위치를 배정합니다.
connect_button.grid(row=0, column=2, padx=5, pady=5)

# 접속대상 윈도우의 폭과 높이 값 설정
width = 280
height = 40

# 현재 접속대상 윈도우가 포함된 운영체제의 화면 해상도
screen_width = win_input.winfo_screenwidth()
screen_height = win_input.winfo_screenheight()

# 접속대상 윈도우를 화면 정가운데 띄우기 위한 x, y 좌표 계산
x = int((screen_width / 2) - (width / 2))
y = int((screen_height / 2) - (height / 2))

# 윈도우의 위치,크기 값 설정
win_input.geometry('%dx%d+%d+%d' % (width, height, x, y))

# 시작시 Entry(입력박스) 위젯에 포커스를 줍니다.
input_addr.focus()

# 윈도우를 띄웁니다.
win_input.mainloop()

        
# tkinter 메인 객체를 생성해 window 변수에 저장합니다.
window = tkinter.Tk()

# 사용자가 접속정보 입력창 윈도우에서 X 버튼을 클릭했을때
# 접속 의사가 없다고 판단하고 프로그램을 종료 하기 위해
# X 버튼 클릭시 발생하는 WM_DELETE_WINDOW 메세지를 window_input_close() 함수와 연결합니다.
window.protocol("WM_DELETE_WINDOW", window_chat_close)


# 윈도우 타이틀을 설정 합니다.
window.title("채팅 클라이언트")

# 채팅 리스트와 입력 박스 + 버튼을 구분하기 위해 프레임을 생성합니다.
frame = tkinter.Frame(window)

# 채팅 메세지 리스트용 스크롤바를
scroll = tkinter.Scrollbar(frame)
scroll.pack(side = tkinter.RIGHT, fill=tkinter.Y)

# 채팅 메세지를 출력할 리스트를 생성하고 스크롤바를 설정합니다.
chat_list = tkinter.Listbox(frame, height=15, width=50, yscrollcommand=scroll.set)

# 스크롤바를 드래그 했을때 chat_list 의 y 축 뷰를 이동하게 설정합니다.
scroll.configure(command=chat_list.yview)

# 채팅리스트 위젯의 배치와 채우기 타입을 설정합니다.
chat_list.pack(side=tkinter.LEFT, fill=tkinter.BOTH, padx=5, pady=5)

# 프레임을 화면에 배치합니다.
frame.pack()

# 입력받은 값을 저장할 String변수
input_msg = tkinter.StringVar()

# window에 속하는 입력박스 생성 후 input_msg 에 변수화 시킵니다.
inputbox = tkinter.Entry(window, textvariable=input_msg)

# inputbox 에서 엔터키를 누르면 send_message 함수를 호출하게 바인딩(연결) 시킵니다.
inputbox.bind("<Return>", send_message)

# 입력박스에 정렬 속성을 주고 화면에 배치(pack) 합니다.
inputbox.pack(side=tkinter.LEFT, fill=tkinter.BOTH, expand=tkinter.YES, padx=5, pady=5)

# window에 속하는 버튼을 생성 하고 send_message 함수와 연결 합니다.
send_button = tkinter.Button(window, text="전송", command = send_message)

# 버튼에 정렬 속성을 주고 화면에 배치 합니다.
send_button.pack(side=tkinter.RIGHT, fill=tkinter.X, padx=5, pady=5)




''' 접속 정보가 입력되었다면 채팅 윈도우 생성 및 동작'''
sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
print("서버 접속시도 [{}:{}]".format(IP, PORT))
sock.connect((IP, PORT))
receive_thread = Thread(target=recv_message)
receive_thread.daemon=True
receive_thread.start()

# 채팅 윈도우의 사이즈 설정
width = 383
height = 292

# 접속대상 윈도우를 화면 정가운데 띄우기 위한 x, y 좌표 계산
x = int((screen_width / 2) - (width / 2))
y = int((screen_height / 2) - (height / 2))

# 윈도우의 위치,크기 값 설정
window.geometry('%dx%d+%d+%d' % (width, height, x, y))

# 윈도우를 실행합니다.
window.mainloop()
```
