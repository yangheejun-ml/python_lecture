## 예전 프로그래밍 스타일은

예를 들어 C/C++ 로 프로그램을 만들려면 생각보다 많은 시스템과 메모리에 대한 지식과 이해가 필요합니다. 

예쁜 그림이 들어간 머그컵을 만든다고 가정을 하면 예전 스타일의 관점으로 보면 우리는 머그컵을 만들기 위해 일단 흙의 성분에 대해 공부해야하고 그걸 구워낼 가마를 지어야 했습니다. 그렇게 어렵사리 흙의 종류를 선택하고 가마를 짓고 나면 이제 머그컵을 만들기 위한 준비가 끝나고 어렵게 한개의 컵을 구워내면 이제 유약이나 색에 대한 공부를 또 해야하는... 그런 인고의 시간을 거쳐서 프로그램을 만들었다고 보면 현재는 그런 과정이 모두 필요없게 그냥 다른사람들이 만들어놓은 수많은 머그컵 재료를 활용해 어떻게 데코레이션 해서 머그컵에 어떤 기능을 넣어 완성시킬까에 대해서만 생각하면 됩니다. 물론 다 알면 좋지만 우리는 그러기에 너무 빠른 시대의 삶을 살고 있습니다. 이런 스타일이 파이썬의 강점인 가장 큰 부분이라고 생각합니다.

# 파이썬

* 1991년 발표된 파이썬은 C 로 구현된 인터프리터식 방식의 프로그래밍 언어 입니다.
* 인터프리터식 이라는 말은 쉽게 컴파일 되지 않고 코드를 한줄씩 내려가며 실행하는 방식입니다.
* 인터프리터식이기 때문에 exe, so 등의 실행파일이 아닌 파이썬 소스 파일인 py 를 python 으로 직접 실행하는 방식입니다.


* 파이썬 소스 코드 [깃헙](https://github.com/python/cpython)
* 파이썬 이스터에그 import this, import antigravity




# 파이썬 장점

1. 가장 간결한 언어 (파이썬 제작자가 가장 의도한 부분) 
   - 코드가 짧고 가독성이 좋음  
   - 생산성이 높음
   - 유연한 언어
    
    
2. 방대한 라이브러리
    - 가장많은 인공지능 라이브러리


물론 이런 장점들은 직접 해보거나 느끼기 전엔 알 수 없는 것들
보통 이런 장점들을 이야기하는 사람들은 전직 개발자들이 자신이 사용하던 언어랑 비교를 하기 때문에 나오는 결론임
처음 프로그래밍을 하는 사람 입장에선 덜 어렵거나 더 어렵거나 어렵기는 마찬가지 입니다.

언어별 네이버 API 호출 예제 [링크](https://developers.naver.com/docs/search/blog/)

# 파이썬 단점

### 속도가 느리다!!
초보에겐 해당사항은 없는 내용입니다만 속도가 느리다는 얘기가 가끔 나오는데 인터프리터적인 언어의 특성상 이미 컴파일되어 바이너리코드화 된 프로그램 보다 느린건 사실일겁니다.

단, 일반적인 분야나 응용 환경속에서 속도가 그정도로 정밀하게 중요한 경우가 많지도 않고 설사 그렇다 하더라도 개발자의 재량에 따라 해당 부분을 C언어 등으로 모듈화해서 제작할 수 있습니다. 또한 그 정도로 고성능의 연산속도가 필요한 프로젝트를 맡아서 진행할 정도라면 이미 실력이 상급이상인 시니어급 개발자라고 봐야하니 그쯤되면 충분히 프로젝트의 문제점을 분석 및 파악하고 다양한 해결책을 테스트할 실력은 될거라 판단됩니다. 


카카오 개발팀에서 작성한 Numpy, C++ Extensions 의 성능 비교 [링크](http://tech.kakao.com/2018/05/15/python-numpy-extensions/)

## 파이썬을 배우는 목적이나 이유

- 개발자, 프로그래머 직업을 꿈꾸는 사람
- 개발자 그런건 관심 없고 아이디어가 있는데 그걸 만들어 보고 싶은 사람
- 아직 아이디어는 없지만 뭔가 서비스를 만들어보고 싶은 사람
- 특별한 이유는 없고 하도 코딩코딩 하니까 막연히 프로그래밍을 배워보고 싶은 사람
- 빅데이터 분석이나 인공지능에 관심 있는 사람
- 개발자는 아니지만 타 업종에 종사는 하고 있지만 뭔가 배우면 도움될께 많은거 같아서

## 파이썬으로 할 수 없는 것

- 모바일 앱 개발 (물론 백앤드 API 는 파이썬으로 만듬)
- 펌웨어 같은 하드웨어 임베디드 프로그램(사물인터넷 라즈베리파이 제외)
- 하드웨어 전용 드라이버 (윈도우 DDK, WDK 같은...)
- 초고성능 연산 프로그램, 고용량 연산(압축-코덱등)의 프로그램(C/C++로 만드는게..)

- 시스템 프로그램

- 웹 프로그래밍

- 문명4, 심즈4, 월드오브탱크, EVE 온라인 등 게임 부분 기능으로 파이썬이 많이 쓰임

- 파이썬으로 할수 있는 걸 모아놓은 깃헙 사이트 https://github.com/vinta/awesome-python

# 진짜 프로그래밍을 배워 개발자가 되려면

요리를 배울려면 음식을 많이 해보면 되고, 화가가 되려면 그림을 많이 그려보면 되고, 피아니스트가 되려면 피아노를 많이 쳐보면 되고....  
그러면 프로그래머가 되려면??? 아마 모두 답을 아실 겁니다.   

요리사가 되려고 꼭 무슨 5성급 호텔에서 나오는 요리를 만들어봐야만 되는건 아닙니다. 요리를 안해보신 분이라면 쌀씻고 밥짓는거 부터 하시겠죠...  라면도 끓여보고 김밥도 싸보고 뭐 이것저것 많이도 만들어보십니다.  


물론 강좌도 보고 공부도 하고 중요합니다만.. 대게 강좌만 보고 공부만 하고 끝내는 분들 많이 봤습니다.  
간단한 프로그램이던 뭐던 프로그램을 많이 만들어보는게 정답입니다. 
똑같은 프로그램이라고 해도 그걸 한번, 두번, 세번, 네번 다시 만들어볼때마다 프로그램은 변화하고 진화하게 되어있습니다.   

근데 막상 프로그램을 만들어 보려고하면 뭘 만들어볼지가 막막하게 됩니다. 사실 이론을 공부하고 이해하는것보다 뭘 만들어보지? 라는 주제를 하나 정하는게 훨씬 더 어려울수도 있습니다.

그래서 저는 개인적으로 이론만큼 중요한게 실습이라고 생각합니다. 제가 앞으로 진행할 강좌는 초반 이론을 진행하고 뒷부분에서는 여러가지 다양한 프로젝트를 100% 실습 위주로 강좌를 진행할 예정입니다.
