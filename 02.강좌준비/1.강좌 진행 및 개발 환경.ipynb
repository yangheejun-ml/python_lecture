# 강좌 진행 및 개발 환경

남박사의 파이썬 실전 100% 활용 강좌는 제일 보편적인 환경인 윈도우10 VisualStudio Code 에서 Python 3.6 로 진행할 예정입니다. 

## OS 별 특징


### 1. 윈도우 10
   - 장점
      - 가장 많이 사용하고 그나마 쓰기 쉬움
      - 많고 다양한 무료 소프트웨어
      - 저렴한 가격. 예전에 비해 아주 많이 저렴해 졌으며 가끔 기회가 좋으면 아주 싼값에 구매 가능
   - 단점
      - 쉬운거 이외의 것을 할려면 알아야 할께 생각보다 많아짐(BIOS, 레지스트리, 서비스, 드라이버.... 등등등)
      - 보통 서비스에서 실서버를 리눅스로 두는 경우가 있는데 이때 윈도우 개발환경과 리눅스 개발환경이 좀 많이 다름
      - 리눅스에서 쉽게 설치되는 라이브러를 윈도우에서 설치할때 애먹는경우가 종종 있음(물론 반대의 상황도 있습니다.)
      - 리눅스와 윈도우를 모두 알아야 하는 문제
      - 실서버를 윈도우서버로 구동할시 비용이 더 많이 듬(라이센스 비용)

### 2. 맥
   - 장점
      - 앱 개발자라면 안드로이드, IOS 모두 개발가능... (하다고는 하나 실제 이런 이유로 구입만해놓고 안하는 사람이 반이상...)
      - 서비스 개발시 실서버를 리눅스로 두는 경우가 많은데 맥 OS가 같은 유닉스 계열이라 개발에 사용한 대부분을 그대로 사용할 수 있음
      - 유닉스 기반 OS 라 터미널 명령에 익숙해지기 좋음
      - 소프트웨어는 애플사가 검증해야 등록되기 때문에 프로그램적인 오류가 별로 없음
   - 단점
      - 애플사의 검증을 받은 소프트웨어라 거의 대부분 오류가 없으나 가끔 오류가 발생하면 해결하기가 너무 힘듬
      - 모든 장점을 덮는 가격
      
### 3. 리눅스
   - 장점
      - 무료
      - 본인의 능력만 있다면 무궁무진한 기회가 있는 환경이며 아예 새로운 운영체제로 변신시킬 수 있음
      - 개발친화적
   - 단점
      - 윈도우나 맥에 비해 불편함
      - 현재는 많이 좋아졌으나 윈도우나 리눅스에 비해 다양한 소프트웨어가 없음

## 어떤 PC 추천?

그냥 본인이 제일 잘 아는거를 쓰거나 주변 지인들이 많이 쓰는 환경을 따라가는게 좋습니다. 물론 지인은 어설프지 않게 컴퓨터를 잘 아는 사람을 기준으로 합니다. 맥을 오래 쓰셨던 분들이라면 그냥 맥 쓰시면 되고 윈도우를 오래 쓰신 분들이라면 윈도우 쓰시면 됩니다.  
각각 장단점이 분명한 환경이라 어떤게 더 좋다고 말하는거 자체가 모순일 수 있습니다.
